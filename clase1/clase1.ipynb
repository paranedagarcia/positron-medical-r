## Medidas de tendencia central en R

```r
# 1. Definición del conjunto de datos (ejemplo: edades de un grupo)
datos = c(14, 15, 22, 25, 22, 21, 24, 24, 30, 25, 22, 36, 40, 35, 36, 25, 22, 28, 22)
boxplot(datos, horizontal = TRUE,
        main = "Diagrama de Caja",
        ylab = "Valores",
        col = "lightblue")

```

```r
# 2. Cálculo de la Media (Promedio aritmético)
# Se utiliza la función mean()
media <- mean(datos)

# 3. Cálculo de la Mediana (Valor central tras ordenar los datos)
mediana <- median(datos)

# 4. Cálculo de la Moda (El valor con mayor frecuencia)
# Creamos una función personalizada para encontrar el valor más frecuente
calcular_moda <- function(v) {
        tab <- table(v)                # Crea una tabla de frecuencias
        indice_max <- which.max(tab)    # Encuentra la posición del valor máximo
        moda <- names(tab)[indice_max]  # Obtiene el nombre (valor) de esa posición
        return(as.numeric(moda))        # Lo devuelve como número
}

moda <- calcular_moda(datos)

# 5. Mostrar los resultados en la consola
cat("--- Medidas de Tendencia Central ---\n")
cat("Datos analizados:", datos, "\n")
cat("Media:", media, "\n")
cat("Mediana:", mediana, "\n")
cat("Moda:", moda, "\n")

# Extra: Resumen estadístico rápido
cat("\n--- Resumen General ---\n")
cat(summary(datos))
```

```r
# grafico
# 1. Define the data vector
data <- c(14, 15, 22, 25, 22, 21, 24, 24, 30, 25, 22, 36, 40, 35, 36, 25, 22, 28, 22)

# 2. Calculate central tendency measures
data_mean   <- mean(data)
data_median <- median(data)

# Helper function for the mode
get_mode <- function(x) {
  ux <- unique(x)
  ux[which.max(tabulate(match(x, ux)))]
}
data_mode <- get_mode(data)

# 3. Create the histogram
# We adjust 'breaks' and 'ylim' for better clarity
hist(data, 
     breaks = 10, 
     col = "lightgray", 
     border = "white",
     main = "Histogram of Data with Central Tendency Measures", 
     xlab = "Value", 
     ylab = "Frequency",
     ylim = c(0, 8))

# 4. Add vertical lines for each measure
# Red solid line represents the Mean
abline(v = data_mean, col = "red", lwd = 3)

# Blue dashed line represents the Median
abline(v = data_median, col = "blue", lwd = 3, lty = 2)

# Green dotted line represents the Mode
abline(v = data_mode, col = "darkgreen", lwd = 3, lty = 3)

# 5. Add a legend to distinguish the lines
legend("topright", 
       legend = c(paste("Mean:", round(data_mean, 2)), 
                  paste("Median:", data_median), 
                  paste("Mode:", data_mode)), 
       col = c("red", "blue", "darkgreen"), 
       lty = 1:3, 
       lwd = 3,
       bg = "white")
```

## Medidas de dispersión

```r
# 1. Definición del vector de datos
# Sustituye este vector con tus valores específicos
datos <- c(14, 15, 22, 25, 22, 21, 24, 24, 30, 25, 22, 36, 40, 35, 36, 25, 22, 28, 22)

# 2. Cálculo de medidas de dispersión

# Rango: La diferencia entre el valor máximo y el mínimo
rango_valor <- diff(range(datos))

# Varianza: Mide la dispersión de los datos respecto a su media
varianza_valor <- var(datos)

# Desviación estándar: La raíz cuadrada de la varianza
desviacion_std <- sd(datos)

# Rango Intercuartílico (IQR): Diferencia entre el tercer y primer cuartil (Q3 - Q1)
rango_iqr <- IQR(datos)

# Coeficiente de variación (CV): Expresa la desviación estándar como porcentaje de la media
# Útil para comparar la variabilidad entre diferentes conjuntos de datos
coef_variacion <- (sd(datos) / mean(datos)) * 100

# 3. Mostrar los resultados en la consola
cat("--- Medidas de Dispersión ---\n")
cat("Rango:                  ", rango_valor, "\n")
cat("Varianza:               ", round(varianza_valor, 2), "\n")
cat("Desviación Estándar:    ", round(desviacion_std, 2), "\n")
cat("Rango Intercuartílico:  ", rango_iqr, "\n")
cat("Coeficiente Variación:  ", round(coef_variacion, 2), "%\n")
```

## Intervalo de confianza

```r
# 1. Define the data vector
data <- c(14, 15, 22, 25, 22, 21, 24, 24, 30, 25, 22, 36, 40, 35, 36, 25, 22, 28, 22)

# 2. Method A: Manual Calculation (Step-by-Step)
n <- length(data)
sample_mean <- mean(data)
sample_sd <- sd(data)
standard_error <- sample_sd / sqrt(n)

# Define confidence level (95%)
conf_level <- 0.95
alpha <- 1 - conf_level

# Find the t-critical value
t_critical <- qt(1 - alpha/2, df = n - 1)

# Calculate the Margin of Error
margin_error <- t_critical * standard_error

# Calculate Bounds
lower_bound <- sample_mean - margin_error
upper_bound <- sample_mean + margin_error

# 3. Method B: Using the built-in t.test function (Shortcut)
ci_result <- t.test(data, conf.level = 0.95)$conf.int

# 4. Display the results
cat("--- Confidence Interval Analysis (95%) ---\n")
cat("Sample Mean:       ", round(sample_mean, 2), "\n")
cat("Standard Error:    ", round(standard_error, 2), "\n")
cat("T-Critical Value:  ", round(t_critical, 3), "\n")
cat("------------------------------------------\n")
cat("95% Confidence Interval: [", round(lower_bound, 2), ", ", round(upper_bound, 2), "]\n")
```
